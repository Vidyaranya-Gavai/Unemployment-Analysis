### The dataset we are going to use contains data on India's unemployment  rate during Covid-19 period which is calculated based on the regions.

```python
# Importing libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import cv2
```

```python
# Importing datset
data = pd.read_csv("F:\\Data Analysis Projects\\Unemployment Analysis\\Unemployment Data.csv")
data
```

```python
data.isnull().sum()
```

#### Some of the columns in the dataset are wrongly named like column containing names of states is named as 'Region' and column containing regions is named as 'Region.1'. We need to modify the column names for better analysis

```python
data.columns = ['States', 'Date', 'Frequency',
                'Estimated Unemployment Rate',
                'Estimated Employees',
                'Estimated Labour Participation Rate',
                'Region', 'Longitude', 'Latitude']
```

#### Now lrt us look at the correlation between the features of this dataset

```python
plt.figure(figsize=(12,10))
plt.rc('font', size=14)
sns.heatmap(data.corr())
plt.show()
```

### Now the final step of our analysis :
#### We are going to see the unemployment rate of each state in India. For this we are going to use 'sunburst' plot.

```python
unemployment = data[['States', 'Region', 'Estimated Unemployment Rate']]
figure = px.sunburst(unemployment, path=['Region', 'States'],
                     values='Estimated Unemployment Rate',
                     width=700, height=700,
                     title='State wise unemployment rate')
figure.show()
```
